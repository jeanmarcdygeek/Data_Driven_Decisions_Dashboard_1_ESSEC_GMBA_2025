```python
import pandas as pd
import json
```

```python
def load_axa_im_fund_json(file_path):
    with open(file_path, "r") as f:
        file_content = json.load(f)
        value_list = file_content["value"]["queries"][0]["results"]
        df = pd.DataFrame(value_list)
        return df
 
```

```python
def load_bnp_im_fund_json(file_path):
    with open(file_path, "r") as f:
        file_content = json.load(f)
        df = pd.DataFrame(file_content)
        return df
```

```python
euro_2 = load_bnp_im_fund_json("em_1.json")
euro_2 = euro_2.sort_values(by="date")
euro_2 = euro_2.pivot(index="date", columns=["index"], values="value").reset_index()
euro_2 = euro_2[euro_2["date"] <= "2024-12-31"]
euro_2["nav"] = euro_2["perf"]

euro_2.to_csv("./data/em_1.csv", index=False)
euro_2
```

```python
euro_1 =  load_axa_im_fund_json("euro_1_data.json")
euro_1 =  euro_1.sort_values(by="date")
euro_1 = euro_1[["date","nav_price"]]
euro_1.columns = ["date","perf"]
euro_1["fund_return"] = euro_1["perf"].pct_change() + 1
euro_1 = euro_1[euro_1["date"] >= "2006-06-12"]
min_date = euro_1["date"].min()
euro_1["nav"] = euro_1["fund_return"]
euro_1.loc[euro_1["date"] == min_date, "nav"] = 100
euro_1["nav"] = euro_1["nav"].cumprod()
euro_1 = euro_1[["date","nav"]].reset_index(drop=True)
bench = euro_2[["date","bench"]]
euro_1 = pd.merge(euro_1, bench, on="date", how="left")
euro_1 = euro_1[euro_1["date"] <= "2024-12-31"]

euro_1.to_csv("./data/euro_1.csv", index=False)
euro_1
```

```python
euro_micro_1 =  load_axa_im_fund_json("euro_micro_1.json")
euro_micro_1 =  euro_micro_1.sort_values(by="date")
euro_micro_1= euro_micro_1[["date","nav_price"]].reset_index(drop=True)
euro_micro_1.columns = ["date","nav"]
euro_micro_1["bench"] = 100

euro_micro_1 = euro_micro_1[euro_micro_1["date"] < "2024-12-31"]
euro_micro_1.to_csv("./data/euro_micro_1.csv", index=False)
euro_micro_1
```
